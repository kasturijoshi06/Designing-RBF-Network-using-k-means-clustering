```python
import numpy as np
import matplotlib.pyplot as plt
import math
import statistics
```

```python
x=np.random.uniform(0,1,size=(100,2))
print (x)
```

```python
d=[]
for i in range (100):
    if (x[i][1] < (1/5)*(np.sin(10*x[i][0]))+0.3) or (((x[i][1]-0.8)**2+(x[i][0]-0.5)**2)<0.15**2):
        d.append(1)
    else:
        d.append(-1)
print (d)
```

```python
c1=[]
c2=[]
for i in range (0,len(d)):
    if d[i]==1:
        c1.append(x[i])
    else:
        c2.append(x[i])
plt.plot(np.array(c1)[:,0],np.array(c1)[:,1],'r+')
plt.plot(np.array(c2)[:,0],np.array(c2)[:,1],'b+')
plt.show()
```

K means for 20 centers

```python
def Euclidean_distance(x1,y1,x2,y2):
    squared_distance =0
    squared_distance= ((x1 - x2)**2) + ((y1 - y2)**2)
    ed = math.sqrt(squared_distance)
    return ed
```

```python
#initializing centers
centroids_pos=[]
centroids_neg=[]

for i in range(10):
    centroids_pos.append(c1[i].tolist())
    print (centroids_pos[i])
centroids_neg=[]
for i in range(10):
    centroids_neg.append(c2[i].tolist())
    print (centroids_neg[i])
```

```python
#positive class
while True:
    c_pos_p=[]
    centroids_pos_new=[]
    for i in range(10):
        c_pos_p.append([])

    print(c_pos_p)
    for i in range(0,len(c1)):
        ed=[]
        for j in range(0,len(centroids_pos)):
            dist = Euclidean_distance(c1[i][0], c1[i][1],centroids_pos[j][0],centroids_pos[j][1])
            ed.append(dist)
        #print(ed)
        min_i=min(ed)
        mi=int(ed.index(min_i))
        c_pos_p[mi].append(c1[i].tolist())
    for k in c_pos_p:
        print(k)
        sumx=0
        sumy=0
        for h in k:
            sumx=sumx+h[0]
            sumy=sumy+h[1]
        avgx=sumx/len(k)
        avgy=sumy/len(k)
        centroids_pos_new.append([avgx,avgy])
    
    if centroids_pos_new == centroids_pos:
        break
    else:
        centroids_pos = centroids_pos_new
```

```python
#Plotiing for positive class
plt.clf()
plt.plot(np.array(centroids_pos)[:,0],np.array(centroids_pos)[:,1],'b+')
plt.plot(np.array(c1)[:,0],np.array(c1)[:,1],'r+')

plt.show()
```

```python
#negative class
while True:
    c_pos_n=[]
    centroids_neg_new=[]
    for i in range(10):
        c_pos_n.append([])

    print(c_pos_n)
    for i in range(0,len(c2)):
        ed=[]
        for j in range(0,len(centroids_neg)):
            dist = Euclidean_distance(c2[i][0], c2[i][1],centroids_neg[j][0],centroids_neg[j][1])
    #         print(dist)
            ed.append(dist)
        #print(ed)
        min_i=min(ed)
        mi=int(ed.index(min_i))
#         print(mi)
        c_pos_n[mi].append(c2[i].tolist())
    for k in c_pos_n:
        print(k)
        sumx=0
        sumy=0
        for h in k:
            sumx=sumx+h[0]
            sumy=sumy+h[1]
        avgx=sumx/len(k)
        avgy=sumy/len(k)
        centroids_neg_new.append([avgx,avgy])
    
    if centroids_neg_new == centroids_neg:
        break
    else:
        centroids_neg = centroids_neg_new
```

```python
#Plot for negative class
plt.clf()
plt.plot(np.array(centroids_neg)[:,0],np.array(centroids_neg)[:,1],'b+')
plt.plot(np.array(c2)[:,0],np.array(c2)[:,1],'r+')

plt.show()
```

```python
plt.clf()
plt.plot(np.array(centroids_pos)[:,0],np.array(centroids_pos)[:,1],'d')
plt.plot(np.array(c1)[:,0],np.array(c1)[:,1],'*')
plt.plot(np.array(centroids_neg)[:,0],np.array(centroids_neg)[:,1],'b+')
plt.plot(np.array(c2)[:,0],np.array(c2)[:,1],'r+')
plt.show()
```

In the figure above, the Diamonds represent centers for positive class and the blue + represent centers for negative class. The yellow stars represent the points belonging to the positive class while the red + represent the points belonging to the negative class.
